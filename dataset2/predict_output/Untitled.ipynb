```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
im_gray = cv2.imread('vgg_unet.png', cv2.IMREAD_GRAYSCALE)
```

```python
(thresh, im_bw) = cv2.threshold(im_gray, 128, 1, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
```

```python
cv2.imwrite('vgg_unet.png', im_bw)
```

```python
plt.imshow(im_bw, cmap='gray')
```
